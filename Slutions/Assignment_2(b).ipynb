##### Building two layer Neural network with hidden layer

```python
#Defining sigmoid function and derivative of sigmoid function

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_der(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
wh = np.random.rand(len(train_X[0]),1000)          # hidden layer's weight
wo = np.random.rand(1000, 1)                       # output layer's weight
lr = 0.5
print("Wh initial: ", wh)
print("\nWo initial: ", wo)

#### Training neural network for 10 epochs

for epoch in range(10):
    # feedforward
    zh = np.dot(train_X, wh)
    ah = sigmoid(zh)

    zo = np.dot(ah, wo)
    ao = sigmoid(zo)

    # Phase1 =======================

    error_out = ((1 / 2) * (np.power((ao - train_Y), 2)))
    print(error_out.sum())

    dcost_dao = ao - train_Y
    dao_dzo = sigmoid_der(zo) 
    dzo_dwo = ah

    dcost_wo = np.dot(dzo_dwo.T, dcost_dao * dao_dzo)

    # Phase 2 =======================

    # dcost_w1 = dcost_dah * dah_dzh * dzh_dw1
    # dcost_dah = dcost_dzo * dzo_dah
    dcost_dzo = dcost_dao * dao_dzo
    dzo_dah = wo
    dcost_dah = np.dot(dcost_dzo , dzo_dah.T)
    dah_dzh = sigmoid_der(zh) 
    dzh_dwh = train_X
    dcost_wh = np.dot(dzh_dwh.T, dah_dzh * dcost_dah)

    # Update Weights ================

    wh -= lr * dcost_wh
    wo -= lr * dcost_wo
```

```python
print("After training: ")
print("Wh: ", wh)
print("\nWo: ", wo)
```

```python
zh = np.dot(test_X[1556], wh)
ah = sigmoid(zh)
zo = np.dot(ah, wo)
ao = sigmoid(zo)
print('predicted value: ',ao)
print('true value: ', test_Y[1556])
if ao>0.5:
    print("It's belong to class 1")
else:
    print("It's belong to class 0")
```

```python
#Function to predict test data with a forward pass
#If 'prediction > 0.5' then it belongs to class'1' otherwise class'0'

def prediction(test_set):
    
    predicted_class = np.zeros(test_set.shape[0], dtype = test_set.dtype)
    
    for i in range(len(test_set)):
        zh = np.dot(test_set[i], wh)
        ah = sigmoid(zh)
        zo = np.dot(ah, wo)
        ao = sigmoid(zo)
        
        if ao > 0.5:
            predicted_class[i] =  1
        else:
            predicted_class[i] = 0
        
    return predicted_class
           
```
